<p style="color:#6a4c93;">
Data cleaning is crucial for model accuracy. In this phase, we perform an initial scan using <b style="color:#2a9d8f;">df.info()</b> and <b style="color:#2a9d8f;">df.describe()</b> to understand data types and statistical distributions.

### 3.2: Handling missing values

```python
# Check for missing values
print("=" * 80)
print("MISSING VALUES CHECK")
print("=" * 80)
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

if missing_values.sum() == 0:
    print("‚úÖ No missing values found!")
```

### 3.3: Categorical value

```python
# Check unique values in categorical columns
print("=" * 80)
print("UNIQUE VALUES IN CATEGORICAL COLUMNS")
print("=" * 80)

categorical_cols = ['Ship Mode', 'Segment', 'Country', 'Region', 'Category', 'Sub-Category']
for col in categorical_cols:
    print(f"\n{col}: {df[col].nunique()} unique values")
    print(df[col].value_counts())
```

### 3.4: Data Preprocessing

```python
# Convert date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Extract year and month for time series analysis
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Year-Month'] = df['Order Date'].dt.to_period('M')

# Calculate Profit Ratio (Profit / Sales * 100)
df['Profit Ratio'] = (df['Profit'] / df['Sales']) * 100

# Handle infinite values (when Sales = 0)
df['Profit Ratio'] = df['Profit Ratio'].replace([np.inf, -np.inf], 0)

print("‚úÖ Data preprocessing completed!")
print(f"\nDate Range: {df['Order Date'].min()} to {df['Order Date'].max()}")
print(f"Total Sales: ${df['Sales'].sum():,.2f}")
print(f"Total Profit: ${df['Profit'].sum():,.2f}")
print(f"Overall Profit Ratio: {(df['Profit'].sum() / df['Sales'].sum()) * 100:.2f}%")
```

### 3.5: Calculate Key Metrics

```python
# Overall metrics
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
overall_profit_ratio = (total_profit / total_sales) * 100

print("=" * 80)
print("KEY BUSINESS METRICS")
print("=" * 80)
print(f"üí∞ Total Sales: ${total_sales:,.2f}")
print(f"üìà Total Profit: ${total_profit:,.2f}")
print(f"üìä Profit Ratio: {overall_profit_ratio:.2f}%")
print(f"üõí Total Orders: {df['Order ID'].nunique():,}")
print(f"üë• Total Customers: {df['Customer ID'].nunique():,}")
print(f"üì¶ Total Products: {df['Product ID'].nunique():,}")
```

### 3.6: Time Series Analysis

```python
# Aggregate data by month
monthly_data = df.groupby('Year-Month').agg({
    'Sales': 'sum',
    'Profit': 'sum'
}).reset_index()

# Calculate monthly profit ratio
monthly_data['Profit Ratio'] = (monthly_data['Profit'] / monthly_data['Sales']) * 100
monthly_data['Date'] = monthly_data['Year-Month'].dt.to_timestamp()

print("üìÖ Monthly aggregation completed!")
monthly_data.head(10)
```

### 4.1: Visualization - Top Metrics Dashboard

### 6.2 Convert dates and extract features
The Order Date column is crucial for time‚Äëbased patterns. We'll extract year, month, day of week, and whether it's a weekend. Also, the time between order and ship might influence sales (e.g., urgent orders could be more expensive).

```python
# 1. Clean column names first
df.columns = df.columns.str.strip()

# 2. Check if columns exist before converting
if 'Order Date' in df.columns and 'Ship Date' in df.columns:
    # Convert to datetime
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    df['Ship Date'] = pd.to_datetime(df['Ship Date'])

    # Extract features
    df['Order Year'] = df['Order Date'].dt.year
    df['Order Month'] = df['Order Date'].dt.month
    df['Order Day'] = df['Order Date'].dt.day
    df['Order DayOfWeek'] = df['Order Date'].dt.dayofweek
    df['Order Weekend'] = (df['Order DayOfWeek'] >= 5).astype(int)

    # Days between order and ship
    df['Ship Days'] = (df['Ship Date'] - df['Order Date']).dt.days

    # Drop columns
    df.drop(['Order Date', 'Ship Date'], axis=1, inplace=True)
    print("Success! Features extracted.")
else:
    print("Columns not found. Available columns are:", df.columns.tolist())
```

### 6.3 Drop irrelevant or high‚Äëcardinality columns
We'll remove columns that are either identifiers or too granular for regression:

`Row ID`, `Order ID`, `Customer ID`, `Customer Name`, `Postal Code`, `Product ID`, `Product Name` (many unique values, likely noisy)

`Country` is constant (all 'United States')

`Profit` is highly correlated with Sales and would cause data leakage (profit is derived from sales and cost). We drop it.

`City` and `State` could be useful, but they have many categories. To keep the model manageable, we'll drop them and rely on `Region`. If you wish, you can try encoding them, but that would increase dimensionality significantly.

```python
cols_to_drop = ['Row ID', 'Order ID', 'Customer ID', 'Customer Name', 'Postal Code',
                'Product ID', 'Product Name', 'Country', 'Profit', 'City', 'State']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```

```python
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']

# One-hot encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

Simultaneously, we perform <b>Hyperparameter Tuning</b> using <b style="color:#2ecc71;">RandomizedSearchCV</b>. By fine-tuning parameters like <i>n_estimators</i> and <i>max_depth</i>, we ensure our models achieve the highest possible <b>R-squared score</b>.
</p>

```python
# Convert to float32 (optional but recommended)
X_train_nn = X_train.values.astype(np.float32)
X_test_nn = X_test.values.astype(np.float32)
y_train_nn = y_train.values.astype(np.float32)
y_test_nn = y_test.values.astype(np.float32)

# Build the model
model_nn = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(X_train_nn.shape[1],)),
    layers.Dropout(0.2),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1)
])

model_nn.compile(optimizer='adam', loss='mse', metrics=['mae'])

early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model_nn.fit(X_train_nn, y_train_nn,
                       validation_split=0.2,
                       epochs=100,
                       batch_size=32,
                       callbacks=[early_stop],
                       verbose=0)
```
